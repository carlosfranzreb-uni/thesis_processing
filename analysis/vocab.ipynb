Analyse the vocabulary.

```python
import json
from collections import Counter
from matplotlib import pyplot as plt
```

```python
vocab = json.load(open('../data/vocab/repo_vocab_1625583660.json'))
```

```python
vocab_cnt = Counter(vocab.values())
len_cnt = Counter([word.count(' ')+1 for word in vocab])
len_cnt.most_common()
```

```python
plt.pie(len_cnt.values(), labels=len_cnt.keys())
plt.title('Number of words per entry.')
plt.show()
```

```python
print(f'There are {len(vocab)} words in the vocabulary.')
print(f'{round(vocab_cnt[1]/len(vocab), 2)} of the words in the vocabulary only occur once in the repositories.')
```

```python
x = [str(i) for i in range(1, 6)] + ['6+']
y = [i[1] for i in vocab_cnt.most_common(5)] + [sum([i[1] for i in vocab_cnt.most_common()][5:])]
plt.pie(y, labels=x)
plt.title('Frequency of each entry.')
plt.show()
```

Now we will look at the words that occur only once.

```python
once = [k for k, v in vocab.items() if v == 1]
once[:10]
```

```python
len_cnt_once = Counter([word.count(' ')+1 for word in once])
len_cnt_once.most_common()
```

```python
plt.pie(len_cnt_once.values(), labels=len_cnt_once.keys())
plt.title('Number of words per entry that occurs once.')
plt.show()
```

```python
for i in range(1, len(len_cnt)+1):
    print(f'{len_cnt_once[i]} out of {len_cnt[i]} entries with {i} words occur only once - {round(len_cnt_once[i]/len_cnt[i], 2)}.')
```
