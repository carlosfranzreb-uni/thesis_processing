Analyse the vocabulary.

```python
import json
from collections import Counter
from matplotlib import pyplot as plt
import numpy as np
```

```python
vocab = json.load(open('../data/vocab/repo_vocab.json'))
vocab_1 = json.load(open('../data/vocab/repo_vocab_step_1.json'))
vocab_2 = json.load(open('../data/vocab/repo_vocab_step_2.json'))
vocab_4 = json.load(open('../data/vocab/repo_vocab_step_4.json'))
vocabs = {
    "Unfiltered": vocab, "Step 1": vocab_1, "Step 2": vocab_2, "Step 4": vocab_4, 
}
```

```python
x = np.array([1, 2, 3, 4])
plt.plot(x, [len(vocab) for vocab in vocabs.values()])
plt.xticks(x, vocabs.keys())
plt.ylabel('No. of entries')
plt.xlabel('Step')
plt.show()
```

```python
width = .2
i = 0
for name, vocab in vocabs.items():
    vocab_cnt = Counter(vocab.values())
    len_cnt = Counter([word.count(' ')+1 for word in vocab])
    values_unordered = len_cnt.most_common()
    values = [t[1] for t in sorted(values_unordered, key=lambda t: t[0])]
    plt.bar(x + (i-3)*(width/2), values, width, label=name)
    i += 2
plt.legend()
plt.xticks(x)
plt.xlabel('No. of words of the entry')
plt.ylabel('No. of entries')
plt.show()
```

```python
print('Average frequency of the entries.')
x, y = [0, 1, 2, 3], []
for name, vocab in vocabs.items():
    y.append(sum(vocab.values()) / len(vocab))
plt.bar(x, y)
plt.xticks(x, vocabs.keys())
plt.ylabel('Avg. frequency')
plt.show()
```

Why does the avg. frequency go down on step 4? Because words with lesser frequency are removed.
