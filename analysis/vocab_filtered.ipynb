Analyse the vocabulary.

```python
import json
from collections import Counter
from matplotlib import pyplot as plt
```

```python
vocab_1 = json.load(open('../data/vocab/repo_vocab_step_1.json'))
vocab_2 = json.load(open('../data/vocab/repo_vocab_step_2.json'))
vocab_3 = json.load(open('../data/vocab/repo_vocab_step_3.json'))
vocab_4 = json.load(open('../data/vocab/repo_vocab_step_4.json'))
vocabs = {
    "Step 1": vocab_1, "Step 2": vocab_2, "Step 3": vocab_3, "Step 4": vocab_4, 
}
```

```python
x = [1, 2, 3, 4]
plt.plot(x, [len(vocab) for vocab in vocabs.values()])
plt.xticks(x)
plt.ylabel('No. of entries')
plt.xlabel('Step')
plt.show()
```

```python
for name, vocab in vocabs.items():
    vocab_cnt = Counter(vocab.values())
    len_cnt = Counter([word.count(' ')+1 for word in vocab])
    values_unordered = len_cnt.most_common()
    values = [t[1] for t in sorted(values_unordered, key=lambda t: t[0])]
    plt.plot(x, values, label=name)
plt.legend()
plt.xticks(x)
plt.xlabel('No. of words')
plt.ylabel('No. of entries')
plt.show()
```

```python
x, y = [], []
for name, vocab in vocabs.items():
    x.append(int(name[-1]))
    y.append(sum(vocab.values()) / len(vocab))
plt.bar(x, y)
plt.xticks(x, vocabs.keys())
plt.ylabel('Avg. frequency')
plt.show()
```
