Analyse the vocabulary.

```python
import json
from collections import Counter
from matplotlib import pyplot as plt
```

```python
vocab = json.load(open('../data/vocab/repo_vocab_step_4.json'))
```

```python
len(vocab)
```

```python
vocab_cnt = Counter(vocab.values())
len_cnt = Counter([word.count(' ')+1 for word in vocab])
len_cnt.most_common()
```

```python
total = [tup[1] for tup in len_cnt.most_common()]
for i in total:
    print(i/sum(total))
```

```python
plt.pie(len_cnt.values(), labels=len_cnt.keys())
plt.title('Number of words per entry.')
plt.show()
```

```python
print(f'There are {len(vocab)} words in the vocabulary.')
print(f'{round(vocab_cnt[1]/len(vocab), 2)} of the words in the vocabulary only occur once in the repositories.')
```

```python
x = [str(i) for i in range(2, 5)] + ['5+']
y = [i[1] for i in vocab_cnt.most_common(3)] + [sum([i[1] for i in vocab_cnt.most_common()][3:])]
plt.pie(y, labels=x)
plt.title('Frequency of each entry.')
plt.show()
```
